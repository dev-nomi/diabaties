#### Step 3.2.2: Removing Punctuation, Numbers, and Special Characters

```python
sample_data['Hobbies'] = sample_data['Hobbies'].str.replace("[^a-zA-Z#]", " ", regex=True)
sample_data['Goals'] = sample_data['Goals'].str.replace("[^a-zA-Z#]", " ", regex=True)
sample_data.head()
```

#### Step 3.2.3 : Removing Short Words

```python
sample_data['Hobbies'] = sample_data['Hobbies'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>5]))
sample_data['Goals'] = sample_data['Goals'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
sample_data.head()
```

# Step 5: Feature Extraction

```python
hobbies = sample_data['Hobbies']
hobbies
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the Count Vectorizer 
count_vectorizer_hobbies = CountVectorizer(
        stop_words = None,  # Stopwords not Removed
        lowercase = True,   # Text Convert into Lower Case
        analyzer = 'word',    # Word n-grams Generation
        token_pattern = r'\w{2,}', #vectorize 2-character words or more
        ngram_range = (1,1), # Word Uni-grams Features
        max_features = None)  # Extract All Features 


# Fit the Count Vectorizer on Input Text
count_vectorizer_hobbies.fit(hobbies)

# Transform the Input Text using Count Vectorizer
transform_features = count_vectorizer_hobbies.transform(hobbies)

# Get the name of Features (Feature  Set)
feature_set = count_vectorizer_hobbies.get_feature_names_out()

# Convert Transformed features into Array and Create a Dataframe
hobbies_input_features = pd.DataFrame(transform_features.toarray(), columns = feature_set)

hobbies_input_features.head()
```

```python
# Save the Trained Count Vectorizer in Pkl File

''' 
*----------------- SAVE_THE_TRAINED_COUNT_VECTORIZER --------------*
|        Function: dump()                                          |
|             Purpose: Save the Trained Model on your Hard Disk    |
|        Arguments:                                                |
|             Model: Model Objects                                 |
|        Return:                                                   |
|             File: Trained Vectorizer will be Saved on Hard Disk  |
*------------------------------------------------------------------* 
'''

pickle.dump(count_vectorizer_hobbies, open('count_vectorizer_hobbies_unigram.pkl', 'wb'))
```

```python
goals = sample_data['Goals']
goals
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the Count Vectorizer 
count_vectorizer_goals = CountVectorizer(
        stop_words = None,  # Stopwords not Removed
        lowercase = True,   # Text Convert into Lower Case
        analyzer = 'word',    # Word n-grams Generation
        token_pattern = r'\w{2,}', #vectorize 2-character words or more
        ngram_range = (1,1), # Word Uni-grams Features
        max_features = None)  # Extract All Features 


# Fit the Count Vectorizer on Input Text
count_vectorizer_goals.fit(goals)

# Transform the Input Text using Count Vectorizer
transform_features = count_vectorizer_goals.transform(goals)

# Get the name of Features (Feature  Set)
feature_set = count_vectorizer_goals.get_feature_names_out()

# Convert Transformed features into Array and Create a Dataframe
goals_input_features = pd.DataFrame(transform_features.toarray(), columns = feature_set)

goals_input_features.head()
```

```python
# Save the Trained Count Vectorizer in Pkl File

''' 
*----------------- SAVE_THE_TRAINED_COUNT_VECTORIZER --------------*
|        Function: dump()                                          |
|             Purpose: Save the Trained Model on your Hard Disk    |
|        Arguments:                                                |
|             Model: Model Objects                                 |
|        Return:                                                   |
|             File: Trained Vectorizer will be Saved on Hard Disk  |
*------------------------------------------------------------------* 
'''

pickle.dump(count_vectorizer_goals, open('count_vectorizer_goals_unigram.pkl', 'wb'))
```

# Step 6: Label Encoding the Sample Data (Output is converted in Numeric Representation)

## Step 6.1: Train the Label Encoder

```python
# Train the Label Encoder

''' 
*------------------ TRAIN_LABEL_ENCODER --------------------*
|        Function: Fit()                                    |
|              Purpose: Fit or Train the Label Encoder      |
|        Arguments:                                         |
|               Labels: Target Values                       |
|        Return:                                            |
|               Instance: Returns an instance of self       |
*-----------------------------------------------------------*
''' 
# Labels

m_group = pd.DataFrame({"M-Group": ["Science ( Bio)","Science (Computer )","O level"]})
i_group = pd.DataFrame({"I-Group": ["FSC (Pre-Engineering)","FSC (Pre - Medical)","ICS (Computer Science)","ICS (Statistics)","A level"]})
program = pd.DataFrame({"Program": ['BS Computer Sciences', 'BS Software Engineering']})

# Initialize the Label Encoders 

m_group_label_encoder = LabelEncoder()
i_group_label_encoder = LabelEncoder()
program_label_encoder = LabelEncoder()

# Train the Label Encoders

m_group_label_encoder.fit(np.ravel(m_group))
i_group_label_encoder.fit(np.ravel(i_group))
program_label_encoder.fit(np.ravel(program))

# Exporting the Label Encoders
pickle.dump(count_vectorizer_goals, open('m_group_label_encoder.pkl', 'wb'))

output = open('m_group_label_encoder.pkl', 'wb')
pickle.dump(m_group_label_encoder, output)
output.close()

output = open('i_group_label_encoder.pkl', 'wb')
pickle.dump(i_group_label_encoder, output)
output.close()

output = open('program_label_encoder.pkl', 'wb')
pickle.dump(program_label_encoder, output)
output.close()
```

## Step 6.2: Label Encode the Output

```python
# Label Encoding of the Output

''' 
*------------------ LABEL_ENCODE_OUTPUT --------------------*
|        Function: Transform()                              |
|              Purpose: Transform Input (Categorical)       |
|                       into Numerical Representation       |
|        Arguments:                                         |
|              Attribute: Target values                     |
|        Return:                                            |
|              Attribute: Numerical Representation          |
*-----------------------------------------------------------*
'''

# Transform Output of into Numerical Representation
print("\n\nCategory Attribute After Label Encoding:")
print("========================================\n")
sample_data["Program"] = program_label_encoder.transform(sample_data['Program'])
print(sample_data["Program"])
```

## Step 6.3: Label Encode the Input

```python
# Label Encoding of the Input

''' 
*------------------- LABEL_ENCODE_INPUT --------------------*
|        Function: Transform()                              |
|              Purpose: Transform Input (Categorical)       |
|                          into Numerical Representation    |
|        Arguments:                                         |
|              Attribute: Target values                     |
|        Return:                                            |
|              Attribute: Numerical Representation          |
*-----------------------------------------------------------*
'''

# Transform Input Attributes into Numerical Representation

print("\n\nM-group Attribute After Label Encoding:")
print("======================================\n")
sample_data["M-Group"] = m_group_label_encoder.transform(sample_data['M-Group'])
print(sample_data["M-Group"])

print("\n\nI-group Attribute After Label Encoding:")
print("======================================\n")
sample_data["I-Group"] = i_group_label_encoder.transform(sample_data['I-Group'])
print(sample_data["I-Group"])
```

# Step 7: Execute the Training Phase 

## Step 7.1: Splitting Sample Data into Training Data and Testing Data

```python
# Splitting Sample Data into Training Data and Testing Data

''' 
*------------------- SPLIT_SAMPLE_DATA ---------------------*
|        Function: train_test_split()                       |
|              Purpose: Split arrays or matrices into       |
|                       random train and test subsets       |
|        Arguments:                                         |
|              arrays: sequence of indexables               |
|              test_size: float or int                      |
|        Return:                                            |
|              splitting: list                              |
*-----------------------------------------------------------*
'''

training_data_encoded, testing_data_encoded = train_test_split( sample_data , test_size=0.2 , random_state=0 , shuffle = False)

# Save the Training and Testing Data into CSV File 

training_data_encoded.to_csv(r'training-data-encoded.csv', index = False, header = True)
testing_data_encoded.to_csv(r'testing-data-encoded.csv', index = False, header = True)

# print Training and Testing Data

print("\n\nTraining Data:")
print("==============\n")
training_data_encoded.head()
```

```python
print("\n\nTesting Data:")
print("==============\n")
testing_data_encoded.head()
```

## Step 7.2: Splitting Input Vectors and Outputs / Labels of Training Data

```python
# Splitting Input Vectors and Outputs / Labels of Training Data

'''
*---------------- SPLIT_INPUT_VECTORS_AND_LABELS --------------*
|        Function: iloc()                                      |
|            Purpose: Splitting Input Vector and Labels        |
|        Arguments:                                            |
|            Attribute: Name or Location Attribute to Split    |
|        Return:                                               |
|            Attribute: Split Attributes                       |
*--------------------------------------------------------------*
'''

print("\n\nInputs Vectors (Feature Vectors) of Training Data:")
print("==================================================\n")
input_vector_train = training_data_encoded.iloc[: , :-1]
input_vector_train.head()
```

```python
print("\n\nOutputs/Labels of Training Data:")
print("================================\n")
print("Program")
output_label_train = training_data_encoded.iloc[: ,-1]
output_label_train
```

## 7.3: Train the GaussianNB Classifier

```python
# Train the GaussianNB Classifier

''' 
*--------------- TRAIN_SUPPORT_VECTOR_CLASSIFIER ------------------*
|       Function: GaussianNB()                                     |
|           Purpose: Train the Algorithm on Training Data          |
|       Arguments:                                                 |
|           Training Data: Provide Training Data to the Model      |
|       Return:                                                    |
|           Parameter: Model return the Training Parameters        |
*------------------------------------------------------------------*
'''

print("\n\nTraining GaussianNB on Training Data")
print("========================================================\n")
# training the model on training set
gnb = GaussianNB()
gnb.fit(input_vector_train,output_label_train)
```

```python
dct = DecisionTreeClassifier(criterion='entropy', random_state=1)
dct.fit(input_vector_train,output_label_train)
```

```python
log_reg = LogisticRegression(random_state=0,solver='lbfgs')
log_reg.fit(input_vector_train,output_label_train)
```

## Step 7.4: Save the Trained Model

```python
# Save the Trained Model

''' 
*--------------------- SAVE_THE_TRAINED_MODEL ---------------------*
|        Function: dump()                                          |
|             Purpose: Save the Trained Model on your Hard Disk    |
|        Arguments:                                                |
|             Model: Model Objects                                 |
|        Return:                                                   |
|             File: Trained Model will be Saved on Hard Disk       |
*------------------------------------------------------------------* 
'''

# Save the Models in a Pkl File

pickle.dump(gnb, open('gnb_trained_model.pkl', 'wb'))
pickle.dump(dct, open('dct_trained_model.pkl', 'wb'))
pickle.dump(log_reg, open('log_reg_trained_model.pkl', 'wb'))
```

# Step 8: Execute the Testing Phase 

```python
# Splitting Input Vectors and Outputs/Labels of Testing Data

'''
*---------------- SPLIT_INPUT_VECTORS_AND_LABELS --------------*
|        Function: iloc()                                      |
|            Purpose: Splitting Input Vector and Labels        |
|        Arguments:                                            |
|            Attribute: Name or Location Attribute to Split    |
|        Return:                                               |
|            Attribute: Split Attributes                       |
*--------------------------------------------------------------*
'''

print("\n\nInputs Vectors (Feature Vectors) of Testing Data:")
print("=================================================\n")
input_vector_test = testing_data_encoded.iloc[: , :-1]
input_vector_test.head()
```

```python
print("\n\nOutputs/Labels of Testing Data:")
print("==============================\n")
print("Program")
output_label_test = testing_data_encoded.iloc[: ,-1]
output_label_test.head()
```

## Step 8.2: Load the Saved Model

```python
# Load the Saved Model

''' 
*------------------- LOAD_SAVED_MODEL --------------------------*
|         Function: load()                                      |
|               Purpose: Method to Load Previously Saved Model  |
|         Arguments:                                            |
|               Model: Trained Model                            |
|         Return:                                               |
|               File: Saved Model will be Loaded in Memory      |
*---------------------------------------------------------------*
'''

# Load the Saved Model

gnb_model = pickle.load(open('gnb_trained_model.pkl', 'rb'))
dct_model = pickle.load(open('dct_trained_model.pkl', 'rb'))
log_reg_model = pickle.load(open('log_reg_trained_model.pkl', 'rb'))
```

## Step 8.3: Evaluate the Machine Learning Model
### Step 8.3.1: Make Predictions with the Trained Models on Testing Data

```python
# Evaluate the Machine Learning Model

''' 
*--------------------- EVALUATE_MACHINE_LEARNING_MODEL ----------------------*
|       Function: Predict()                                                  |
|             Purpose: Make a Prediction using Algorithm on Test Data        |
|       Arguments:                                                           |
|            Testing Data: Provide Test data to the Trained Model            |
|       Return:                                                              |
|            Predictions: Model return Predictions                           |
*----------------------------------------------------------------------------* 
'''

# Provide Test data to the Trained Model

gnb_model_predictions = gnb_model.predict(input_vector_test)
dct_model_predictions = dct_model.predict(input_vector_test)
log_reg_model_predictions = log_reg_model.predict(input_vector_test)
testing_data_encoded.copy(deep=True)
pd.options.mode.chained_assignment = None
testing_data_encoded["gnb_Predictions"] = gnb_model_predictions
testing_data_encoded["dct_Predictions"] = dct_model_predictions
testing_data_encoded["log_reg_Predictions"] = log_reg_model_predictions

# Save the Predictions into CSV File

testing_data_encoded.to_csv(r'model-predictions.csv', index = False, header = True)

model_predictions = testing_data_encoded 
print("\n\nPredictions Returned by models:")
print("==========================================\n")
model_predictions.head()
```

## Step 8.4: Calculate the Accuracy Score

```python
# Calculate the Accuracy Score

''' 
/*------------------------ CALCULATE_ACCURACY_SCORE -------------------*
|          Function: accuracy_score()                                  |
|                Purpose: Evaluate the algorithm on Testing data       |
|          Arguments:                                                  |
|                Prediction: Predicted values                          |
|                Label: Actual values                                  |
|          Return:                                                     |
|                Accuracy: Accuracy Score                              |
*----------------------------------------------------------------------*
'''

# Calculate the Accuracy

gnb_model_accuracy_score = accuracy_score(model_predictions["Program"],model_predictions["gnb_Predictions"])
dct_model_accuracy_score = accuracy_score(model_predictions["Program"],model_predictions["dct_Predictions"])
log_reg_model_accuracy_score = accuracy_score(model_predictions["Program"],model_predictions["log_reg_Predictions"])

print("\n\nAccuracy Score by GaussianNB:")
print("===============\n")
print(round(gnb_model_accuracy_score,2))

print("\n\nAccuracy Score by DecisionTreeClassifier:")
print("===============\n")
print(round(dct_model_accuracy_score,2))

print("\n\nAccuracy Score by LogisticRegression:")
print("===============\n")
print(round(log_reg_model_accuracy_score,2))
```

## Step 8.5: Calculate the F1 Score

```python
# Calculate the F1 Score

''' 
/*------------------------ CALCULATE_ACCURACY_SCORE -------------------*
|          Function: f1_score()                                        |
|                Purpose: Evaluate the algorithm on Testing data       |
|          Arguments:                                                  |
|                Prediction: Predicted values                          |
|                Label: Actual values                                  |
|          Return:                                                     |
|                F1 Score                                              |
*----------------------------------------------------------------------*
'''

# Calculate the f1 score

gnb_model_f1_score = f1_score(model_predictions["Program"],model_predictions["gnb_Predictions"])
dct_model_f1_score = f1_score(model_predictions["Program"],model_predictions["dct_Predictions"])
log_reg_model_f1_score = f1_score(model_predictions["Program"],model_predictions["log_reg_Predictions"])

print("\n\nF1 Score by GaussianNB:")
print("===============\n")
print(round(gnb_model_f1_score,2))

print("\n\nF1 Score by DecisionTreeClassifier:")
print("===============\n")
print(round(dct_model_f1_score,2))

print("\n\nF1 Score by LogisticRegression:")
print("===============\n")
print(round(log_reg_model_f1_score,2))
```

## Step 8.6: Results

```python
models = ['GaussianNB','DecisionTreeClassifier','LogisticRegression']

accuracy = [gnb_model_accuracy_score,dct_model_accuracy_score,log_reg_model_accuracy_score]
f1_score = [gnb_model_f1_score,dct_model_f1_score,log_reg_model_f1_score]

compare = pd.DataFrame({'Model':models,'Accuracy':accuracy,'F1_Score':f1_score},index=[i for i in range(1,4)])

compare.T
```

```python
plt.figure(figsize=(18,5))

ax = plt.gca()

compare.plot(kind='line',x='Model',y='Accuracy',ax=ax)
compare.plot(kind='line',x='Model',y='F1_Score', color='red', ax=ax)

plt.show()
```

# Step 9: Execute the Application Phase

## Step 9.1: Take Input from User

```python
# Take Input from User

''' 
*---------------- TAKE_USER_INPUT ----------------*
'''

hobbies_input = input("\nPlease enter hobbies here: ").strip()
goals_input = input("\nPlease enter your goals here: ").strip()
m_group_input = input("\nPlease enter your Matric Group ( Science ( Bio), Science (Computer) , O level) : ").strip()
m_english_marks_input = input("\nPlease enter M-English Marks here: ").strip()
m_urdu_marks_input = input("\nPlease enter M-Urdu Marks here: ").strip()
m_islamic_studies_marks_input = input("\nPlease enter M-Islamic Studies Marks here: ").strip()
m_pak_studies_marks_input = input("\nPlease enter M-Pak Studies Marks here: ").strip()
m_bio_comp_marks_input = input("\nPlease enter M-Bio/Comp Marks here: ").strip()
m_physics_marks_input = input("\nPlease enter M-Physics Marks here: ").strip()
m_chemistry_marks_input = input("\nPlease enter M-Chemistry Marks here: ").strip()
m_mathematics_marks_input = input("\nPlease enter M-Mathematics Marks here: ").strip()

i_group_input = input("\nPlease enter your Inter Group ( FSC (Pre-Engineering) , FSC (Pre - Medical), ICS (Computer Science), ICS (Statistics), A level ) : ").strip()
i_english_marks_input = input("\nPlease enter I-English Marks here: ").strip()
i_urdu_marks_input = input("\nPlease enter I-Urdu Marks here: ").strip()
i_islamic_studies_marks_input = input("\nPlease enter I-Islamic Studies Marks here: ").strip()
i_pak_studies_marks_input = input("\nPlease enter I-Pak Studies Marks here: ").strip()
i_math_bio_marks_input = input("\nPlease enter I-Math/Bio Marks here: ").strip()
i_physics_statistics_marks_input = input("\nPlease enter I-Physics/Statistics here: ").strip()
i_comp_chem_marks_input = input("\nPlease enter I-Comp/Chem Marks here: ").strip()
```

## Step 9.2: Convert User Input into Feature Vector (Exactly Same as Feature Vectors of Sample Data)

```python
# Convert User Input into Feature Vector

user_input = pd.DataFrame({ 
    'Hobbies': [hobbies_input],
    'Goals': [goals_input],
    "M-Group": [m_group_input],
    "M-English Marks":[m_english_marks_input],
    "M-Urdu Marks":[m_urdu_marks_input],
    "M-Islamic Studies Marks":[m_islamic_studies_marks_input],
    "M-Pak Studies Marks":[m_pak_studies_marks_input],
    "M-Bio/Comp Marks":[m_bio_comp_marks_input],
    "M-Physics Marks":[m_physics_marks_input],
    "M-Chemistry Marks":[m_chemistry_marks_input],
    "M-Mathematics Marks":[m_mathematics_marks_input],
    "I-Group": [i_group_input],
    "I-English Marks":[i_english_marks_input],
    "I-Urdu Marks":[i_urdu_marks_input],
    "I-Pak Studies Marks":[i_pak_studies_marks_input],
    "I-Islamic Studies Marks":[i_islamic_studies_marks_input],
    "I-Math/Bio":[i_math_bio_marks_input],
    "I-Physics/Statistics":[i_physics_statistics_marks_input],
    "I-Comp/Chem":[i_comp_chem_marks_input],
})

print("\n\nUser Input Feature Vector:")
print("==========================\n")
user_input
```

## Step 9.3: Label Encoding of Feature Vector (Exactly Same as Label Encoded Feature Vectors of Sample Data)

```python
# Label Encoding

''' 
*----------------- TRANSFORM_UNSEEN_INTPUT_FEATURES ---------------*
|           Function: Transform()                                  |
|                 Purpose: Transform Input (Categorical) into      |
|                          Numerical Representation                |
|           Arguments:                                             |
|                 Attribute: Target values                         |
|           Return:                                                |
|                 Attribute: Numerical Representation              |
*------------------------------------------------------------------*   
'''

# Transform Input (Categorical) Attributes of Unseen Data into Numerical Representation

unseen_data_features = user_input.copy()
unseen_data_features["M-Group"] = m_group_label_encoder.transform(user_input['M-Group'])
unseen_data_features["I-Group"] = i_group_label_encoder.transform(user_input['I-Group'])

print("\n\nUser Input Encoded Feature Vector:")
print("==================================\n")
unseen_data_features
```

```python
unseen_data_features['Hobbies'] = unseen_data_features['Hobbies'].str.replace("[^a-zA-Z#]", " ", regex=True)
unseen_data_features['Goals'] = unseen_data_features['Goals'].str.replace("[^a-zA-Z#]", " ", regex=True)
unseen_data_features
```

```python
unseen_data_features['Hobbies'] = unseen_data_features['Hobbies'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>5]))
unseen_data_features['Goals'] = unseen_data_features['Goals'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))
unseen_data_features
```

```python
tokenized_hobbies = unseen_data_features['Hobbies'].apply(lambda x: x.split())
tokenized_hobbies
```

```python
tokenized_goals = unseen_data_features['Goals'].apply(lambda x: x.split())
tokenized_goals
```

```python
tokenized_hobbies = tokenized_hobbies.apply(lambda x: [ps.stem(i) for i in x])
tokenized_hobbies
```

```python
tokenized_goals = tokenized_hobbies.apply(lambda x: [ps.stem(i) for i in x])
tokenized_goals
```

```python
unseen_data_features['Hobbies'] = ' '.join(tokenized_hobbies[0])
unseen_data_features.head()
```

```python
unseen_data_features['Goals'] = ' '.join(tokenized_goals[0])
unseen_data_features.head()
```

```python
hobbies = unseen_data_features['Hobbies']
hobbies
```

```python
# Transform the Input Text using Count Vectorizer
transform_features = count_vectorizer_hobbies.transform(hobbies)

# Get the name of Features (Feature  Set)
feature_set = count_vectorizer_hobbies.get_feature_names_out()

# Convert Transformed features into Array and Create a Dataframe
hobbies_input_features = pd.DataFrame(transform_features.toarray(), columns = feature_set)

hobbies_input_features.head()
```

```python
goals = unseen_data_features['Goals']
goals
```

```python
# Transform the Input Text using Count Vectorizer
transform_features = count_vectorizer_goals.transform(goals)

# Get the name of Features (Feature  Set)
feature_set = count_vectorizer_goals.get_feature_names_out()

# Convert Transformed features into Array and Create a Dataframe
goals_input_features = pd.DataFrame(transform_features.toarray(), columns = feature_set)

goals_input_features.head()
```

```python
unseen_data_features.drop(['Hobbies', 'Goals'], axis=1, inplace=True)
unseen_data_features
```

```python
unseen_data_features = pd.concat([unseen_data_features, hobbies_input_features], axis=1, join='inner')
unseen_data_features.head()
```

```python
unseen_data_features = pd.concat([unseen_data_features, goals_input_features], axis=1, join='inner')
unseen_data_features.head()
```

```python
# Load the Saved Model

''' 
*----------------------- LOAD_SAVED_MODEL --------------------------*
|         Function: load()                                          |
|             Purpose: Method to Load Previously Saved Model        |
|         Arguments:                                                |
|               Model: Trained Model                                |
|         Return:                                                   |
|               File: Saved Model will be Loaded in Memory          |
*-------------------------------------------------------------------*
'''

# Load the Saved Model

model = pickle.load(open('gnb_trained_model.pkl', 'rb'))
```

```python
# Prediction of Unseen Instance

''' 
*----------------------------  MODEL_PREDICTION --------------------------*
|           Function: predict()                                           |
|                 Purpose: Use Trained Model to Predict the Output        |
|                          of Unseen Instances                            |
|           Arguments:                                                    |
|                 User Data: Label Encoded Feature Vector of              |
|                            Unseen Instances                             |
|           Return:                                                       |
|                 Survival:  Computer Science or Software Engineering     |
*-------------------------------------------------------------------------*
'''

# Make a Prediction on Unseen Data

predicted_survival = model.predict(unseen_data_features)

if(predicted_survival == 0): 
    prediction = "Computer Science"
if(predicted_survival == 1):
    prediction = "Software Engineering"

# Add the Prediction in a Pretty Table

pretty_table = PrettyTable()
pretty_table.add_column("       ** Prediction **       ",[prediction])
print(pretty_table)
```
